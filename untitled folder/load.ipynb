```python
import pandas as pd

# Đường dẫn tới file JSON
file_path = "paula_choice_full_details.json"

# Đọc file JSON vào DataFrame
df = pd.read_json(file_path)

# Hiển thị 5 dòng đầu tiên
print(df.head())
```

```python
import json

# Kiểm tra cấu trúc JSON
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# In ra 5 phần tử đầu tiên để kiểm tra
print(data[:5])

# Nếu cần, chuyển đổi JSON lồng nhau thành DataFrame
if isinstance(data, list):
    df = pd.json_normalize(data)
    print(df.head())
else:
    print("JSON không phải là danh sách các dictionary.")
```
